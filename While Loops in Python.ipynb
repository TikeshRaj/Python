```python
# While Loops in Python
```

```python
number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        continue
    number = number + 1
```

```python
number = 0

while number < 5:
    number = number + 1
    print(number)
    if number == 3:
        continue
    
```

```python
number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No Longer > 5')
```
